```python
import pandas as pd
```

```python
df1 = pd.read_excel('train.xlsx')
df1.head()
```

```python
df1_test = pd.read_excel('test.xlsx')
df1_test.head()
```

```python
# Define a function to map values
def map_values(value):
    if value in [1, 2]:
        return "negative"
    elif value == 3:
        return "neutral"
    elif value in [4, 5]:
        return "positive"
    else:
        return value

# Apply the function to the column
df1['new_rating'] = df1['rating'].apply(map_values)
df1_test['new_rating'] = df1_test['rating'].apply(map_values)
df1.to_csv('train.csv',index = False)
df1_test.to_csv('test.csv',index = False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer



# Sample data
X_train = df1['title_hindi'] # train text data
y_train = df1['new_rating']  # training target classes

X_test = df1_test['title_hindi']# test text data
y_test = df1_test['new_rating'] # test target classes
# Preprocessing: Convert text to numerical features using TF-IDF
# Create a TfIDFVectorizer instance for unigram features
vectorizer = TfidfVectorizer()

# Fit and transform the training data
X_train_unigram = vectorizer.fit_transform(X_train)

# Transform the test data using the same vectorizer
X_test_unigram = vectorizer.transform(X_test)

# Initialize SVC
svc_unigram = SVC()

# Train the SVC model
svc_unigram.fit(X_train_unigram, y_train)

# Predict on the test set
y_pred_unigram = svc_unigram.predict(X_test_unigram)

# Calculate accuracy
accuracy_unigram = accuracy_score(y_test, y_pred_unigram)
print("Accuracy (Unigram) using TFidf:", accuracy_unigram)
print(classification_report(y_test, y_pred_unigram))

# Create a TfidfVectorizer instance for bigram features
vectorizer_bigram = TfidfVectorizer(ngram_range=(2, 2))

# Fit and transform the training data
X_train_bigram = vectorizer_bigram.fit_transform(X_train)

# Transform the test data using the same vectorizer
X_test_bigram = vectorizer_bigram.transform(X_test)

# Initialize SVC
svc_bigram = SVC()

# Train the SVC model
svc_bigram.fit(X_train_bigram, y_train)

# Predict on the test set
y_pred_bigram = svc_bigram.predict(X_test_bigram)

# Calculate accuracy
accuracy_bigram = accuracy_score(y_test, y_pred_bigram)
print("Accuracy (Bigram) using TFidf:", accuracy_bigram)
print(classification_report(y_test, y_pred_bigram))
```
