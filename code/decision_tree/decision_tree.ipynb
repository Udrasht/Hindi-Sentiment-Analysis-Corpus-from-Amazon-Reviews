```python
import pandas as pd
```

```python
df1 = pd.read_excel('train.xlsx')
df1.head()
```

```python
df1_test = pd.read_excel('test.xlsx')
df1_test.head()
```

```python
# Define a function to map values
def map_values(value):
    if value in [1, 2]:
        return "negative"
    elif value == 3:
        return "neutral"
    elif value in [4, 5]:
        return "positive"
    else:
        return value

# Apply the function to the column
df1['new_rating'] = df1['rating'].apply(map_values)
df1_test['new_rating'] = df1_test['rating'].apply(map_values)
```

```python
df1.head()
```

```python
df1_test.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report

X_train_unigram = df1['title_hindi']
y_train = df1['new_rating']
X_test_unigram = df1_test['title_hindi']
y_test = df1_test['new_rating']



unigram_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(analyzer='word', ngram_range=(1, 1))),
    ('clf', DecisionTreeClassifier())
])

# Fit the model on the training data
unigram_pipeline.fit(X_train_unigram, y_train)

# Predict on the test data
unigram_predictions = unigram_pipeline.predict(X_test_unigram)

# Calculate accuracy
unigram_accuracy = accuracy_score(y_test, unigram_predictions)
print("Accuracy for unigram using TFidf:", unigram_accuracy)
print(classification_report(y_test,unigram_predictions))

bigram_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(analyzer='word', ngram_range=(2, 2))),
    ('clf', DecisionTreeClassifier())
])

# Fit the model on the training data
bigram_pipeline.fit(df1['title_hindi'], df1['new_rating'])

# Predict on the test data
bigram_predictions = bigram_pipeline.predict(df1_test['new_rating'])

# Calculate accuracy
bigram_accuracy = accuracy_score(y_test, bigram_predictions)
print("Accuracy for bigram using TFidf:", bigram_accuracy)
print(classification_report(y_test,bigram_predictions))
```
