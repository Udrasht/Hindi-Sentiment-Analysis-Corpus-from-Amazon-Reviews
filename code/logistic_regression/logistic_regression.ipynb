```python
import pandas as pd
```

```python
df1 = pd.read_excel('train.xlsx')
df1.head()
```

```python
df1_test = pd.read_excel('test.xlsx')
df1_test.head()
```

```python
# Define a function to map values
def map_values(value):
    if value in [1, 2]:
        return "negative"
    elif value == 3:
        return "neutral"
    elif value in [4, 5]:
        return "positive"
    else:
        return value

# Apply the function to the column
df1['new_rating'] = df1['rating'].apply(map_values)
df1_test['new_rating'] = df1_test['rating'].apply(map_values)
```

```python
df1.head()
```

```python
df1_test.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report


X_train = df1['title_hindi']
y_train = df1['new_rating']
X_test = df1_test['title_hindi']
y_test = df1_test['new_rating']

# Define pipelines for unigram and bigram features
unigram_pipeline = make_pipeline(TfidfVectorizer(ngram_range=(1, 1)), LogisticRegression(max_iter=1000))
bigram_pipeline = make_pipeline(TfidfVectorizer(ngram_range=(2, 2)), LogisticRegression(max_iter=1000))

# Fit and predict using unigram features
unigram_pipeline.fit(X_train, y_train)
unigram_predictions = unigram_pipeline.predict(X_test)
unigram_accuracy = accuracy_score(y_test, unigram_predictions)

print("Accuracy for unigram using Tfidf:", unigram_accuracy)
print(classification_report(y_test,unigram_predictions))

# Fit and predict using bigram features
bigram_pipeline.fit(X_train, y_train)
bigram_predictions = bigram_pipeline.predict(X_test)
bigram_accuracy = accuracy_score(y_test, bigram_predictions)

print("Accuracy for bigram using Tfidf:", bigram_accuracy)
print(classification_report(y_test,bigram_predictions))
```
